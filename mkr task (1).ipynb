```python
data['no_employee']=data.no_employees.map({'Jun-25':'6-25','01-May':'1-5','More than 1000':'1000+','26-100':'26-100','100-500':'100-500','500-1000':'500-1000'})
data['gender']=data.Gender.map({'Female':'F','female':'F','f':'F','F':'F','Male':'M','male':'M','m':'M','M':'M'})
```

work interfere can be a result to the number of employees around, so we use the statistical method to fill the missing values in work interfere with the average of no of employees. so we put the condition that if the no of employees are more,there is a chance of more work interference.
no of employee-0: 1-5
1:6-25
2:26-100
3:100-500
4:500-1000
5:more than 1000

```python
def interfere(cols):
    work_interfere=cols[0]
    no_employee=cols[1]
    
    if pd.isnull(work_interfere):
        
        if no_employee=='1-5':
            return 'Never'
        elif no_employee=='6-25':
            return 'Rarely'
        elif no_employee=='26-100':
            return 'Rarely'
        elif no_employee=='100-500':
            return 'Sometimes'
        elif no_employee=='500-1000':
            return 'Sometimes'
        else:
            return 'Often'
    else:
        return work_interfere
```

```python
data['work_interfere']=data[['work_interfere','no_employee']].apply(interfere,axis=1)
```
